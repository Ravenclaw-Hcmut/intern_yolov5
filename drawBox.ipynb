```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import random
import matplotlib.patches as patches
from PIL import Image, ImageDraw
import argparse
```

```python
image_path = 'datasets/data_th/images/val/THM_255.jpg'
label_path = 'datasets/data_th/labels/val/THM_255.txt'

def get_coords(x_center, y_center, box_width, box_height, image_width, image_height):
    # class x_center y_center width height
    x1 = image_width * (x_center - box_width / 2)
    x2 = image_width * (x_center + box_width / 2)
    
    y1 = image_height * (y_center - box_height / 2)
    y2 = image_height * (y_center + box_height / 2)
    
    return [x1, y1, x2, y2]


def draw_bounding_boxs(_img_path, _img_label):
    fl = open(_img_label, 'r')
    coords = fl.readlines()
    fl.close()
            
            
    im = Image.open(_img_path)
    draw = ImageDraw.Draw(im)
    
    print(im.size)
    
    for dt in coords:            
        # class x_center y_center width height
        dt_split = dt.split(' ')
        dt_split_int = list(map(lambda x: float(x), dt_split))
        # print(dt_split_int)
                
        _, tmp_x_cen, tmp_y_cen, tmp_box_w, tmp_box_h = dt_split_int
        tmp_xy1xy2 = get_coords(tmp_x_cen, tmp_y_cen, tmp_box_w, tmp_box_h, (im.size)[0], (im.size)[1])
        # print(tmp_xy1xy2)
        draw.rectangle(tmp_xy1xy2, outline=(0, 0, 0, 255), width=10)
        
    im.save('im2h28.jpg')
    im.show()

print(type(get_coords(0.5,0.5,0.5,0.5,100,100)))
print(get_coords(0.5,0.5,0.5,0.5,100,100))
```

```python

draw_bounding_boxs(image_path, label_path)
```

### draw (RGBA)

```python
from PIL import Image, ImageDraw

# size of image
canvas = (400, 300)

# scale ration
scale = 1
thumb = canvas[0]/scale, canvas[1]/scale

# rectangles (width, height, left position, top position)
frames = [(50, 50, 5, 5), (60, 60, 100, 50), (100, 100, 205, 120)]

# init canvas
im = Image.new('RGBA', canvas, (255, 255, 255, 255))
draw = ImageDraw.Draw(im)

# draw rectangles
for frame in frames:
    x1, y1 = frame[2], frame[3]
    x2, y2 = frame[2] + frame[0], frame[3] + frame[1]
    draw.rectangle([x1, y1, x2, y2], outline=(0, 0, 0, 255))

# make thumbnail
im.thumbnail(thumb)

im = im.convert('RGB')

# save image
im.save('im.jpg')
```
